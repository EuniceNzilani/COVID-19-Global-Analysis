```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from datetime import datetime
```

```python
# Set display options for better readability
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# We'll directly load the dataset from Our World in Data URL
print("1. Loading COVID-19 data...")
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
df = pd.read_csv(url)
```

```python
# Data Loading & Exploration
# --------------------------

print("\n2. Exploring the dataset...")

# Print the shape of the dataset
print(f"Shape of the dataset: {df.shape}")

# Display the first 5 rows
print("\nFirst 5 rows:")
print(df.head())

# Display column names
print("\nColumn names:")
print(df.columns.tolist())

# Display summary statistics for numerical columns
print("\nSummary statistics:")
print(df.describe())

# Count of missing values per column
print("\nMissing values count:")
print(df.isnull().sum())
```

```python
# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# 3️⃣ Data Cleaning
# ----------------
print("\n3. Cleaning the data...")
```

```python
# Select countries of interest
countries_of_interest = ['United States', 'India', 'Brazil', 'United Kingdom', 'Russia', 'France', 'Germany', 'South Africa', 'Kenya', 'Japan']
df_countries = df[df['location'].isin(countries_of_interest)].copy()
```

```python

# Check for missing values in key columns
print("\nMissing values in key columns for selected countries:")
key_columns = ['total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_vaccinations', 'people_vaccinated']
df_countries[key_columns].isnull().sum()
```

```python
# Forward fill missing values for numeric data within same country
print("\nFilling missing values...")
for country in countries_of_interest:
    country_mask = df_countries['location'] == country
    df_countries.loc[country_mask, key_columns] = df_countries.loc[country_mask, key_columns].fillna(method='ffill')
```

```python
# 4️⃣ Exploratory Data Analysis (EDA)
# ---------------------------------
print("\n4. Performing Exploratory Data Analysis...")

# Filter out data only until the latest available date with data
latest_date = df_countries['date'].max()
print(f"Latest date in dataset: {latest_date}")

# Set up a figure for total cases over time
plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    country_data = df_countries[df_countries['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.title('Total COVID-19 Cases Over Time by Country', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Cases', fontsize=12)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Using log scale for better visualization (optional)
plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    country_data = df_countries[df_countries['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.title('Total COVID-19 Cases Over Time by Country (Log Scale)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Cases (log scale)', fontsize=12)
plt.legend()
plt.grid(True)
plt.yscale('log')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Total deaths over time
plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    country_data = df_countries[df_countries['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)

plt.title('Total COVID-19 Deaths Over Time by Country', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Deaths', fontsize=12)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Calculate death rate
df_countries['death_rate'] = (df_countries['total_deaths'] / df_countries['total_cases'] * 100).replace([np.inf, -np.inf], np.nan)

# Plot death rate
plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    country_data = df_countries[df_countries['location'] == country]
    plt.plot(country_data['date'], country_data['death_rate'], label=country)

plt.title('COVID-19 Death Rate Over Time by Country (Deaths/Cases %)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Death Rate (%)', fontsize=12)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Daily new cases - 7-day rolling average for smoother visualization
plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    country_data = df_countries[df_countries['location'] == country].copy()
    country_data['rolling_new_cases'] = country_data['new_cases'].rolling(window=7).mean()
    plt.plot(country_data['date'], country_data['rolling_new_cases'], label=country)

plt.title('Daily New COVID-19 Cases (7-day Rolling Average)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('New Cases (7-day avg)', fontsize=12)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Top 10 countries by total cases (bar chart)
latest_data = df.groupby('location').last().reset_index()
top_cases = latest_data.sort_values('total_cases', ascending=False).head(10)

plt.figure(figsize=(14, 8))
sns.barplot(x='total_cases', y='location', data=top_cases)
plt.title('Top 10 Countries by Total COVID-19 Cases', fontsize=16)
plt.xlabel('Total Cases', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 5️⃣ Visualizing Vaccination Progress
# ----------------------------------
print("\n5. Analyzing vaccination progress...")

# Total vaccinations over time
plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    country_data = df_countries[df_countries['location'] == country]
    # Skip countries with no vaccination data
    if not country_data['total_vaccinations'].isna().all():
        plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)

plt.title('Total COVID-19 Vaccinations Over Time by Country', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Vaccinations', fontsize=12)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Vaccination percentage (people_vaccinated_per_hundred)
plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    country_data = df_countries[df_countries['location'] == country]
    # Skip countries with no vaccination data
    if not country_data['people_vaccinated_per_hundred'].isna().all():
        plt.plot(country_data['date'], country_data['people_vaccinated_per_hundred'], label=country)

plt.title('Percentage of Population Vaccinated Over Time by Country', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Percentage Vaccinated (%)', fontsize=12)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Comparison of full vs partial vaccination (stacked bar chart)
latest_vacc_data = latest_data.dropna(subset=['people_fully_vaccinated_per_hundred', 'people_vaccinated_per_hundred'])
latest_vacc_data = latest_vacc_data.sort_values('people_vaccinated_per_hundred', ascending=False).head(10)

# Create data for partially and fully vaccinated
latest_vacc_data['partially_vaccinated'] = latest_vacc_data['people_vaccinated_per_hundred'] - latest_vacc_data['people_fully_vaccinated_per_hundred']

plt.figure(figsize=(14, 8))
bar_width = 0.8
bars = plt.barh(latest_vacc_data['location'], latest_vacc_data['people_fully_vaccinated_per_hundred'],
         label='Fully Vaccinated', alpha=0.8, color='darkblue')
bars2 = plt.barh(latest_vacc_data['location'], latest_vacc_data['partially_vaccinated'], left=latest_vacc_data['people_fully_vaccinated_per_hundred'],
          label='Partially Vaccinated', alpha=0.6, color='lightblue')

plt.title('COVID-19 Vaccination Status (Top 10 Countries by Vaccination Rate)', fontsize=16)
plt.xlabel('Percentage of Population (%)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.legend()
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```

```python
# 6️⃣ Choropleth Map Visualization
# ------------------------------
print("\n6. Creating a global choropleth map...")

# Get latest data for each country
latest_global_data = df.groupby('location').last().reset_index()

# Create a choropleth map for total cases per million
fig = px.choropleth(
    latest_global_data,
    locations="iso_code",
    color="total_cases_per_million",
    hover_name="location",
    color_continuous_scale="Viridis",
    title="COVID-19 Total Cases per Million by Country"
)
fig.show()

# Another choropleth for vaccination percentage
fig = px.choropleth(
    latest_global_data,
    locations="iso_code",
    color="people_vaccinated_per_hundred",
    hover_name="location",
    color_continuous_scale="YlGnBu",
    title="COVID-19 Vaccination Percentage by Country"
)
fig.show()
```

```python
# 7️⃣ Insights & Reporting
# ----------------------
print("\n7. Generating insights...")

# Calculate key metrics
latest_global_data['death_rate'] = (latest_global_data['total_deaths'] / latest_global_data['total_cases'] * 100).replace([np.inf, -np.inf], np.nan)

# Top 5 countries by total cases
top_cases_countries = latest_global_data.sort_values('total_cases', ascending=False).head(5)
print("\nTop 5 countries by total COVID-19 cases:")
top_cases_countries[['location', 'total_cases', 'total_deaths', 'death_rate']]

# Top 5 countries by vaccination percentage (among those with data)
vacc_data = latest_global_data.dropna(subset=['people_vaccinated_per_hundred'])
top_vacc_countries = vacc_data.sort_values('people_vaccinated_per_hundred', ascending=False).head(5)
print("\nTop 5 countries by vaccination percentage:")
top_vacc_countries[['location', 'people_vaccinated_per_hundred']]

# Countries with highest and lowest death rates (minimum 1000 cases)
significant_outbreaks = latest_global_data[latest_global_data['total_cases'] > 1000].copy()
high_death_rate = significant_outbreaks.sort_values('death_rate', ascending=False).head(5)
low_death_rate = significant_outbreaks.sort_values('death_rate').head(5)

print("\nCountries with highest death rates (among countries with >1000 cases):")
high_death_rate[['location', 'total_cases', 'total_deaths', 'death_rate']]

print("\nCountries with lowest death rates (among countries with >1000 cases):")
low_death_rate[['location', 'total_cases', 'total_deaths', 'death_rate']]
```


## Key Insights from COVID-19 Data Analysis

1. **Global Impact Variation**: The data shows significant variation in COVID-19 impact across countries, with some experiencing much higher case and death rates than others.

2. **Vaccination Disparities**: Vaccination rollout speeds varied considerably between nations, with some achieving high vaccination rates much faster than others.

3. **Death Rate Factors**: Death rates appear to correlate with various factors such as healthcare system capacity, demographic structures, and timing of interventions.

4. **Infection Waves**: There were distinct waves of infections visible in the time series data, often corresponding to the emergence of new variants.

5. **Vaccination Effectiveness**: Countries with earlier and more aggressive vaccination programs generally showed better containment of cases and deaths in later stages of the pandemic.

This analysis provides a snapshot of the global COVID-19 situation, but further investigation into specific regions, healthcare policies, and socioeconomic factors would yield deeper insights.
